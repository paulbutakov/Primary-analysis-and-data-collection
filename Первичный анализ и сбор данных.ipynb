````python
# –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ –∫–∞–∂–¥–æ–π —Ç–∞–±–ª–∏—Ü–µ
print("Orders - –¥—É–±–ª–∏–∫–∞—Ç—ã –ø–æ order_id:", orders['order_id'].duplicated().sum())
print("Users - –¥—É–±–ª–∏–∫–∞—Ç—ã –ø–æ user_id:", users['user_id'].duplicated().sum())
print("Events - –¥—É–±–ª–∏–∫–∞—Ç—ã –ø–æ event_id:", events['event_id'].duplicated().sum())
print("Campaign_costs - –ø–æ–ª–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã:", campaign_costs.duplicated().sum())

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –ø–æ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –ø–æ–ª–µ–π
print("\n–î—É–±–ª–∏–∫–∞—Ç—ã Orders –ø–æ –∫–ª—é—á–µ–≤—ã–º –ø–æ–ª—è–º:", 
      orders.duplicated(subset=['order_id', 'user_id', 'order_date']).sum())
print("–î—É–±–ª–∏–∫–∞—Ç—ã Events –ø–æ –∫–ª—é—á–µ–≤—ã–º –ø–æ–ª—è–º:", 
      events.duplicated(subset=['event_id', 'user_id', 'event_date', 'event_type']).sum())
````

````python
# –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
def check_missing_data(df, df_name):
    print(f"\n{df_name}:")
    missing = df.isnull().sum()
    missing_percent = (missing / len(df)) * 100
    missing_df = pd.DataFrame({
        '–ü—Ä–æ–ø—É—â–µ–Ω–æ': missing,
        '–ü—Ä–æ—Ü–µ–Ω—Ç': missing_percent
    })
    return missing_df[missing_df['–ü—Ä–æ–ø—É—â–µ–Ω–æ'] > 0]

print(check_missing_data(orders, "Orders"))
print(check_missing_data(users, "Users"))
print(check_missing_data(events, "Events"))
print(check_missing_data(campaign_costs, "Campaign_costs"))
````

–í—ã—Ä—É—á–∫–∞ –±–∏–∑–Ω–µ—Å–∞ ‚Äî 5% –∫–æ–º–∏—Å—Å–∏–∏ –æ—Ç —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –ª—é–±–æ–π –ø—Ä–æ–¥–∞–∂–∏.

–ü–æ—Å–º–æ—Ç—Ä–∏—Ç–µ –Ω–∞ –æ–±—â—É—é –¥–∏–Ω–∞–º–∏–∫—É –∫–ª—é—á–µ–≤—ã—Ö –º–µ—Ç—Ä–∏–∫: –Ω–∞–ø—Ä–∏–º–µ—Ä, Revenue, ARPU –∏–ª–∏ ARPPU, MAU, WAU, DAU, AOV, Retention, CTR, CR –≤ –∑–∞–∫–∞–∑. –°–ª–µ–¥–æ–≤–∞—Ç—å —ç—Ç–æ–º—É —Å–ø–∏—Å–∫—É –Ω–µ–æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ, –∏—Å—Ö–æ–¥–∏—Ç–µ –∏–∑ –¥–∞–Ω–Ω—ã—Ö.

–í–∞—à–∞ —Ü–µ–ª—å ‚Äî –±—ã—Å—Ç—Ä–æ –ø–æ–ª—É—á–∏—Ç—å —Ü–µ–ª–æ—Å—Ç–Ω—É—é –∫–∞—Ä—Ç–∏–Ω—É, –Ω–µ –ø—Ä–∏–º–µ–Ω—è—è —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—é –∏ –Ω–µ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—è –≥–ª—É–±–æ–∫–∏–µ —Å—Ä–µ–∑—ã. –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –ª–∏ –ø—Ä–æ–¥—É–∫—Ç –±–∞–∑–æ–≤—ã–º –æ–∂–∏–¥–∞–Ω–∏—è–º, –≤–∏–¥–µ–Ω –ª–∏ —Ä–æ—Å—Ç, –µ—Å—Ç—å –ª–∏ —Ç—Ä–µ–≤–æ–∂–Ω—ã–µ —Å–∏–≥–Ω–∞–ª—ã.

–†–µ–∑—É–ª—å—Ç–∞—Ç–æ–º –∑–∞–¥–∞–Ω–∏—è –±—É–¥–µ—Ç –≤–∞—à –≤—ã–≤–æ–¥ –æ –¥–∏–Ω–∞–º–∏–∫–µ –æ—Å–Ω–æ–≤–Ω—ã—Ö –ø—Ä–æ–¥—É–∫—Ç–æ–≤—ã—Ö –º–µ—Ç—Ä–∏–∫.

````python
# –í—ã—Ä—É—á–∫–∞ (5% –∫–æ–º–∏—Å—Å–∏—è)
orders['revenue'] = orders['total_price'] * 0.05

# –ï–∂–µ–º–µ—Å—è—á–Ω–∞—è –≤—ã—Ä—É—á–∫–∞
monthly_revenue = orders.set_index('order_date').resample('M')['revenue'].sum()
print(f"\n–û–±—â–∞—è –≤—ã—Ä—É—á–∫–∞: {orders['revenue'].sum():,.0f} —Ä—É–±.")
print(f"–°—Ä–µ–¥–Ω–µ–º–µ—Å—è—á–Ω–∞—è –≤—ã—Ä—É—á–∫–∞: {monthly_revenue.mean():,.0f} —Ä—É–±.")

# AOV (Average Order Value)
aov = orders.groupby('order_id')['total_price'].sum().mean()
print(f"–°—Ä–µ–¥–Ω–∏–π —á–µ–∫ (AOV): {aov:,.0f} —Ä—É–±.")
````

````python
# –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ –º–µ—Ç—Ä–∏–∫–∏ (DAU, WAU, MAU)
def calculate_active_users(events, period):
    if period == 'D':
        return events.groupby(events['event_date'].dt.date)['user_id'].nunique()
    elif period == 'W':
        return events.groupby(events['event_date'].dt.to_period('W'))['user_id'].nunique()
    else:  # M
        return events.groupby(events['event_date'].dt.to_period('M'))['user_id'].nunique()

dau = calculate_active_users(events, 'D')
wau = calculate_active_users(events, 'W')
mau = calculate_active_users(events, 'M')

print("–ê–∫—Ç–∏–≤–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏:")
print(f"–°—Ä–µ–¥–Ω–∏–π DAU: {dau.mean():.0f}")
print(f"–°—Ä–µ–¥–Ω–∏–π WAU: {wau.mean():.0f}")
print(f"–°—Ä–µ–¥–Ω–∏–π MAU: {mau.mean():.0f}")

# –ù–æ–≤—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏
new_users_monthly = users.groupby(users['registration_date'].dt.to_period('M')).size()
print(f"\n–ù–æ–≤—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –≤ –º–µ—Å—è—Ü: {new_users_monthly.mean():.0f}")
````

````python
# CTR (Click-Through Rate)
product_clicks = events[events['event_type'] == 'product_click'].shape[0]
product_views = events[events['event_type'] == 'product_view'].shape[0]
ctr = product_clicks / product_views if product_views > 0 else 0
print(f"CTR (–∫–ª–∏–∫–∏/–ø—Ä–æ—Å–º–æ—Ç—Ä—ã): {ctr:.2%}")

# Conversion Rate
unique_buyers = orders['user_id'].nunique()
unique_viewers = events[events['event_type'] == 'product_view']['user_id'].nunique()
cr_to_purchase = unique_buyers / unique_viewers if unique_viewers > 0 else 0
print(f"Conversion Rate (–ø–æ–∫—É–ø–∞—Ç–µ–ª–∏/—Å–º–æ—Ç—Ä–µ–≤—à–∏–µ —Ç–æ–≤–∞—Ä—ã): {cr_to_purchase:.2%}")
````

````python
# –ú–æ–Ω–µ—Ç–∏–∑–∞—Ü–∏—è (ARPU/ARPPU)
total_revenue = orders['revenue'].sum()
total_users = users['user_id'].nunique()
paying_users = orders['user_id'].nunique()
total_customers = orders['user_id'].nunique()

arpu = total_revenue / total_users
arppu = total_revenue / paying_users

print("–ú–æ–Ω–µ—Ç–∏–∑–∞—Ü–∏—è:")
print(f"ARPU: {arpu:,.0f} —Ä—É–±.")
print(f"ARPPU: {arppu:,.0f} —Ä—É–±.")
print(f"Paying users ratio: {paying_users/total_users:.2%}")
````

````python
print("–î–ò–ù–ê–ú–ò–ö–ê –ú–ï–¢–†–ò–ö:")
print("–ï–∂–µ–º–µ—Å—è—á–Ω–∞—è –≤—ã—Ä—É—á–∫–∞:")
for month, revenue in monthly_revenue.items():
    print(f" {month.strftime('%Y-%m')}: {revenue:,.0f} —Ä—É–±.")

print("\n–ê–∫—Ç–∏–≤–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –ø–æ –º–µ—Å—è—Ü–∞–º:")
for month, active in mau.items():
    print(f" {month}: {active} –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π")
````

````python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# –í—ã—Ä—É—á–∫–∞
axes[0,0].plot(monthly_revenue.index, monthly_revenue.values, marker='o')
axes[0,0].set_title('–ï–∂–µ–º–µ—Å—è—á–Ω–∞—è –≤—ã—Ä—É—á–∫–∞')
axes[0,0].set_ylabel('–†—É–±–ª–µ–π')

# –ê–∫—Ç–∏–≤–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏
axes[0,1].plot(range(len(mau)), mau.values, marker='o', color='orange')
axes[0,1].set_title('Monthly Active Users (MAU)')
axes[0,1].set_ylabel('–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')

# –ù–æ–≤—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏
axes[1,0].bar(range(len(new_users_monthly)), new_users_monthly.values, color='green', alpha=0.7)
axes[1,0].set_title('–ù–æ–≤—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –ø–æ –º–µ—Å—è—Ü–∞–º')
axes[1,0].set_ylabel('–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')

# AOV
monthly_aov = orders.groupby(orders['order_date'].dt.to_period('M'))['total_price'].mean()
axes[1,1].plot(range(len(monthly_aov)), monthly_aov.values, marker='s', color='red')
axes[1,1].set_title('–°—Ä–µ–¥–Ω–∏–π —á–µ–∫ (AOV) –ø–æ –º–µ—Å—è—Ü–∞–º')
axes[1,1].set_ylabel('–†—É–±–ª–µ–π')

plt.tight_layout()
plt.show()
````

- –†–µ–∫–æ–º–µ–Ω–¥—É—é —Ç–∞–∫–∂–µ —Ä–∞–∑–¥–µ–ª–∏—Ç—å –∞–Ω–∞–ª–∏–∑ –Ω–∞ –¥–≤–∞ –ø–µ—Ä–∏–æ–¥–∞: "—Ä–æ—Å—Ç" (—è–Ω–≤–∞—Ä—å-–∏—é–ª—å) –∏ "—Å–Ω–∏–∂–µ–Ω–∏–µ" (–∞–≤–≥—É—Å—Ç-–¥–µ–∫–∞–±—Ä—å) –¥–ª—è –±–æ–ª–µ–µ —Ç–æ—á–Ω–æ–≥–æ –ø–æ–Ω–∏–º–∞–Ω–∏—è —Ñ–∞–∫—Ç–æ—Ä–æ–≤, –≤–ª–∏—è—é—â–∏—Ö –Ω–∞ –∫–∞–∂–¥—É—é —Ñ–∞–∑—É.  –£ –Ω–∞—Å —á—ë—Ç–∫–æ –≤–∏–¥–Ω–∞ —Ç–æ—á–∫–∞ –ø–µ—Ä–µ–ª–æ–º–∞ –≤ –∏—é–ª–µ. –≠—Ç–æ –Ω–µ –ø—Ä–æ—Å—Ç–æ "—Å–µ–∑–æ–Ω–Ω—ã–π —Ñ–∞–∫—Ç–æ—Ä", –∞ –≤–µ—Ä–æ—è—Ç–Ω–æ –∏–∑–º–µ–Ω–µ–Ω–∏–µ —Å—Ç—Ä–∞—Ç–µ–≥–∏–∏ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è. –¢—É—Ç —É –Ω–∞—Å –ø—Ä–æ–±–ª–µ–º–∞ –≥–ª—É–±–∂–µ - –Ω–æ–≤—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –ø–æ—Å–ª–µ –∏—é–ª—è –∫–∞—á–µ—Å—Ç–≤–µ–Ω–Ω–æ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –æ—Ç —Ä–∞–Ω–Ω–∏—Ö. –≠—Ç–æ –Ω–µ –ø—Ä–æ—Å—Ç–æ "—Å–Ω–∏–∂–µ–Ω–∏–µ –≤–æ–≤–ª–µ—á—ë–Ω–Ω–æ—Å—Ç–∏", –∞ —Ñ—É–Ω–¥–∞–º–µ–Ω—Ç–∞–ª—å–Ω–æ–µ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –∫–∞—á–µ—Å—Ç–≤–∞ —Ç—Ä–∞—Ñ–∏–∫–∞.

- –ú–æ–∂–Ω–æ –¥–æ–±–∞–≤–∏—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –ø—Ä–æ–≤–µ—Ä–∫—É –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ —Ç—Ä–µ–Ω–¥–∞ - –∏—Å–ø–æ–ª—å–∑—É–π linear regression –¥–ª—è –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏—è, —á—Ç–æ —Å–Ω–∏–∂–µ–Ω–∏–µ –Ω–µ —Å–ª—É—á–∞–π–Ω–æ: `scipy.stats.linregress(range(len(mau)), mau['MAU'])`.

### 3. –û—Ü–µ–Ω–∫–∞ –º–µ—Ç—Ä–∏–∫ –º–æ–Ω–µ—Ç–∏–∑–∞—Ü–∏–∏ –∏ —é–Ω–∏—Ç-—ç–∫–æ–Ω–æ–º–∏–∫–∏

–í—ã–ø–æ–ª–Ω–∏—Ç–µ –∞–Ω–∞–ª–∏–∑ –º–æ–Ω–µ—Ç–∏–∑–∞—Ü–∏–∏. –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, —Å–∫–æ–ª—å–∫–æ –∑–∞—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –ø—Ä–æ–¥—É–∫—Ç –∏ —Å–∫–æ–ª—å–∫–æ —Ç—Ä–∞—Ç–∏—Ç –Ω–∞ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏–µ.

–í–∞–∂–Ω—ã –∫–∞–∫ –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏ (ARPU, AOV), —Ç–∞–∫ –∏ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ (CAC, ROI, Payback Period).

–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–π—Ç–µ –ø–µ—Ä–≤—ã–µ –≤—ã–≤–æ–¥—ã –ø–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –∞–Ω–∞–ª–∏–∑–∞ –º–µ—Ç—Ä–∏–∫. –°–¥–µ–ª–∞–π—Ç–µ —ç—Ç–æ –≤ —Ñ–æ—Ä–º–∞—Ç–µ –∑–∞–ø–∏—Å–∫–∏ –¥–ª—è –∑–∞–∫–∞–∑—á–∏–∫–∞. –ò–∑–ª–æ–∂–∏—Ç–µ –≤—Å–µ –æ—Å–Ω–æ–≤–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ç–∞–∫, —á—Ç–æ–±—ã –ø—Ä–æ—Å–ª–µ–∂–∏–≤–∞–ª–∞—Å—å —Å–≤—è–∑—å –º–µ–∂–¥—É –º—ã—Å–ª—è–º–∏. –ù–µ –ø–∏—à–∏—Ç–µ –º–Ω–æ–≥–æ ‚Äî –∑–∞–∫–∞–∑—á–∏–∫ –Ω–µ –±—É–¥–µ—Ç —á–∏—Ç–∞—Ç—å –∑–∞–ø–∏—Å–∫—É –ø–æ–ª–Ω–æ—Å—Ç—å—é, –∞ –ø—Ä–æ—Å—Ç–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–∏—Ç.

–ù–∞–ø–∏—à–∏—Ç–µ –≤—ã–≤–æ–¥—ã –≤ –∫–æ–Ω—Ü–µ –±–ª–æ–∫–∞, —á—Ç–æ–±—ã –∑–∞–∫–∞–∑—á–∏–∫ –º–æ–≥ –±—ã—Å—Ç—Ä–æ –æ–∑–Ω–∞–∫–æ–º–∏—Ç—å—Å—è —Å –æ—Å–Ω–æ–≤–Ω—ã–º–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏.

````python
# –ê–Ω–∞–ª–∏–∑ –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º –∏ –¥–µ–º–æ–≥—Ä–∞—Ñ–∏–∏
region_analysis = users.merge(
    orders.groupby('user_id')['total_price'].sum().reset_index(),
    on='user_id', 
    how='left'
).fillna(0)

# –ü–æ —Ä–µ–≥–∏–æ–Ω–∞–º
region_stats = region_analysis.groupby('region').agg({
    'user_id': 'count',
    'total_price': ['sum', 'mean'],
    'age': 'mean'
}).round(0)

print("–¢–æ–ø-5 —Ä–µ–≥–∏–æ–Ω–æ–≤ –ø–æ –≤—ã—Ä—É—á–∫–µ:")
print(region_stats.nlargest(5, ('total_price', 'sum')))

# –ü–æ –≤–æ–∑—Ä–∞—Å—Ç—É –∏ –ø–æ–ª—É
demographic_stats = region_analysis.groupby(['gender', pd.cut(region_analysis['age'], bins=[0, 25, 35, 45, 100])]).agg({
    'user_id': 'count',
    'total_price': 'mean'
}).round(0)

print("\n–î–µ–º–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑:")
print(demographic_stats)
````

````python
# –ê–Ω–∞–ª–∏–∑ –≤–æ—Ä–æ–Ω–∫–∏ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏

# –°–æ–±–∏—Ä–∞–µ–º –≤—Å–µ —ç—Ç–∞–ø—ã –≤–æ—Ä–æ–Ω–∫–∏
funnel_stages = {
    '–í—Å–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏': users['user_id'].nunique(),
    '–ü–æ—Å–µ—Ç–∏–ª–∏ —Å–∞–π—Ç': events['user_id'].nunique(),
    '–ü—Ä–æ—Å–º–æ—Ç—Ä–µ–ª–∏ —Ç–æ–≤–∞—Ä': events[events['event_type'] == 'product_view']['user_id'].nunique(),
    '–î–æ–±–∞–≤–∏–ª–∏ –≤ –∫–æ—Ä–∑–∏–Ω—É': events[events['event_type'] == 'add_to_cart']['user_id'].nunique(),
    '–ù–∞—á–∞–ª–∏ –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ': events[events['event_type'] == 'checkout_start']['user_id'].nunique(),
    '–ó–∞–≤–µ—Ä—à–∏–ª–∏ –ø–æ–∫—É–ø–∫—É': orders['user_id'].nunique()
}

funnel_df = pd.DataFrame({
    'stage': list(funnel_stages.keys()),
    'users': list(funnel_stages.values())
})
funnel_df['conversion_rate'] = (funnel_df['users'] / funnel_df['users'].iloc[0] * 100).round(2)
funnel_df['dropoff_rate'] = (1 - funnel_df['users'] / funnel_df['users'].shift(1)) * 100
funnel_df['dropoff_rate'] = funnel_df['dropoff_rate'].fillna(0).round(2)

print("–í–æ—Ä–æ–Ω–∫–∞ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏:")
print(funnel_df)
````

````python
# –ê–Ω–∞–ª–∏–∑ —Ç–æ–≤–∞—Ä–Ω—ã—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π
category_analysis = orders.groupby('category_name').agg({
    'order_id': 'count',
    'quantity': 'sum',
    'total_price': 'sum',
    'unit_price': 'mean'
}).round(0).sort_values('total_price', ascending=False)

print("–¢–æ–ø –∫–∞—Ç–µ–≥–æ—Ä–∏–π –ø–æ –≤—ã—Ä—É—á–∫–µ:")
print(category_analysis.head(8))

# –ê–Ω–∞–ª–∏–∑ –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –ø–æ–∫—É–ø–æ–∫ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
repeat_purchases = orders.groupby(['user_id', 'category_name']).size().reset_index(name='purchase_count')
repeat_rate_by_category = repeat_purchases[repeat_purchases['purchase_count'] > 1].groupby('category_name').size()

print(f"\n–ö–∞—Ç–µ–≥–æ—Ä–∏–∏ —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –ø–æ–∫—É–ø–æ–∫:")
print(repeat_rate_by_category.nlargest(5))
````

- **–†–∞–∑–º–µ—Ä—ã –≥—Ä—É–ø–ø:** –æ–ø—Ä–µ–¥–µ–ª—è—é—Ç—Å—è –ø–æ —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–æ–º—É —Ç—Ä–∞—Ñ–∏–∫—É TikTok –∑–∞ –ø–µ—Ä–∏–æ–¥ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞  
- **–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π:** –∑–∞–ø—É—Å–∫ –±–µ–∑ –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞—Å—á—ë—Ç–∞ –º–æ—â–Ω–æ—Å—Ç–∏ –∏ MDE –∏–∑-–∑–∞ —Å—Ä–æ—á–Ω–æ—Å—Ç–∏ –∑–∞–¥–∞—á–∏

````python
# –ê–Ω–∞–ª–∏–∑ –æ—Å–Ω–æ–≤–Ω—ã—Ö –º–µ—Ç—Ä–∏–∫ - –∫–æ–Ω–≤–µ—Ä—Å–∏—è –≤ –ø–µ—Ä–≤—É—é –ø–æ–∫—É–ø–∫—É

# –û–±—ä–µ–¥–∏–Ω—è–µ–º –¥–∞–Ω–Ω—ã–µ –æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è—Ö, –≥—Ä—É–ø–ø–∞—Ö –∏ –∑–∞–∫–∞–∑–∞—Ö
users_with_groups = users_ab.merge(ab_split, on='user_id', how='inner')
users_with_orders = users_with_groups.merge(
    orders_ab[['user_id', 'order_date']].drop_duplicates(), 
    on='user_id', 
    how='left',
    indicator=True
)

# –ü–æ–º–µ—á–∞–µ–º, —Å–æ–≤–µ—Ä—à–∏–ª –ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –ø–æ–∫—É–ø–∫—É
users_with_orders['made_purchase'] = (users_with_orders['_merge'] == 'both').astype(int)

# –°—á–∏—Ç–∞–µ–º –∫–æ–Ω–≤–µ—Ä—Å–∏—é –ø–æ –≥—Ä—É–ø–ø–∞–º
conversion_by_group = users_with_orders.groupby('group')['made_purchase'].agg(['count', 'sum', 'mean'])
conversion_by_group['conversion_rate'] = (conversion_by_group['mean'] * 100).round(2)
conversion_by_group = conversion_by_group.rename(columns={
    'count': 'total_users',
    'sum': 'buyers',
    'mean': 'conversion'
})

print("–ö–æ–Ω–≤–µ—Ä—Å–∏—è –≤ –ø–µ—Ä–≤—É—é –ø–æ–∫—É–ø–∫—É –ø–æ –≥—Ä—É–ø–ø–∞–º:")
print(conversion_by_group[['total_users', 'buyers', 'conversion_rate']])

# –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –∑–Ω–∞—á–∏–º–æ—Å—Ç—å
group_control = users_with_orders[users_with_orders['group'] == 'control']['made_purchase']
group_treatment = users_with_orders[users_with_orders['group'] == 'treatment']['made_purchase']

t_stat, p_value = stats.ttest_ind(group_control, group_treatment)
print(f"\n–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞:")
print(f"T-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ (p < 0.05): {p_value < 0.05}")
````

```python
from scipy import stats
import numpy as np

def bootstrap_mean_diff(group1, group2, n_bootstrap=10000):
    differences = []
    for _ in range(n_bootstrap):
        sample1 = np.random.choice(group1, size=len(group1), replace=True)
        sample2 = np.random.choice(group2, size=len(group2), replace=True)
        differences.append(sample2.mean() - sample1.mean())
    
    ci_lower = np.percentile(differences, 2.5)
    ci_upper = np.percentile(differences, 97.5)
    
    return ci_lower, ci_upper
```
